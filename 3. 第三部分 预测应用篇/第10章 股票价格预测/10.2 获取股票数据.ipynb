这里以贵州茅台股票为例，说明pandas_datareader库中股票数据的获取方法及简单的可视化

```python
import pandas as pd
import pandas_datareader.data as web
import datetime as dt
data = web.DataReader('600519.ss','yahoo', dt.datetime(2019,8,1),dt.datetime(2019,8,31))
data.head()
```

```python
kldata=data.values[:,[2,3,1,0]] # 分别对应开盘价、收盘价、最低价和最高价
from pyecharts import options as opts
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts.charts import Kline

kobj = Kline().add_xaxis(data.index.strftime("%Y-%m-%d").tolist()).add_yaxis("贵州茅台-日K线图", kldata.tolist()).set_global_opts(
            yaxis_opts=opts.AxisOpts(is_scale=True),
            xaxis_opts=opts.AxisOpts(is_scale=True),
            title_opts=opts.TitleOpts(title=""))

kobj.load_javascript()
```

```python
kobj.render_notebook()
```
